```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_train = pd.read_csv("D:\\Skills\\ACM Sigchi\\Week 1\\Bonus\\house-prices-advanced-regression-techniques\\train.csv")
df_test = pd.read_csv("D:\\Skills\\ACM Sigchi\\Week 1\\Bonus\\house-prices-advanced-regression-techniques\\test.csv")

df = pd.concat([df_train, df_test], axis=0)
df.head()
```

```python
X = df.iloc[:, :-1]
y_train = df_train.iloc[:, -1]
```

```python
columnsTodrop = ['PoolQC', 'MiscFeature', 'Alley', 'Fence']
X.drop(columns=columnsTodrop, inplace=True)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
X_imputed = imputer.fit_transform(X)
```

```python
X_imputed = pd.DataFrame(X_imputed, columns=X.columns)
```

```python
for column in X_imputed.select_dtypes(include=['object']).columns:
    try:
        X_imputed[column] = pd.to_numeric(X_imputed[column])
    except ValueError:
        pass
```

```python
X_imputed.head()
```

```python
X_numerical = X_imputed.select_dtypes(include=['int64', 'float64'])
X_categorical = X_imputed.select_dtypes(object)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in X_categorical.columns:
    X_categorical[column] = encoder.fit_transform(X_categorical[column])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_numerical_scaled = scaler.fit_transform(X_numerical)
```

```python
X_numerical_scaled = pd.DataFrame(df_numerical_scaled, columns=X_numerical.columns)
```

```python
X_numerical_scaled.head()
```

```python
X_processed = pd.concat([X_numerical_scaled, X_categorical], axis=1)
X_processed.drop(columns=['Id'], inplace=True)
```

```python
X_train = np.array(X_processed[0:df_train.shape[0]])
X_test = np.array(X_processed[df_train.shape[0]:])
```

```python
y_train = np.array(y_train)
```

```python
from sklearn.model_selection import train_test_split

X_to_train, X_valid, y_to_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_to_train, y_to_train)

y_pred = model.predict(X_valid)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100)
model.fit(X_to_train, y_to_train)

y_pred = model.predict(X_valid)
```

```python
from sklearn.metrics import r2_score

print(f"R2 Score = {r2_score(y_true=y_valid, y_pred=y_pred)}")
```

```python
from sklearn.svm import SVR

model = SVR(kernel='poly')
model.fit(X_to_train, y_to_train)

y_pred = model.predict(X_valid)
```

```python
from sklearn.metrics import r2_score

print(f"R2 Score = {r2_score(y_true=y_valid, y_pred=y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

param_grids = {
    'Linear Regression': {},
    'Random Forest Regressor': {
        'n_estimators': [50, 100, 200]
    },
    'Support Vector Regressor': {
        'kernel': ['linear', 'rbf']
    },
    'Gradient Boosting Regressor': {
        'n_estimators': [100, 200]
    },
    'XGBoost Regressor': {
        'n_estimators': [100, 200]
    }
}

models = {
    'Linear Regression': LinearRegression(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Support Vector Regressor': SVR(),
    'Gradient Boosting Regressor': GradientBoostingRegressor(),
    'XGBoost Regressor': XGBRegressor(eval_metric='mlogloss')
}
for model_name, model in models.items():
    param_grid = param_grids[model_name]
    
    if not param_grid:
        model.fit(X_to_train, y_to_train)
        y_pred = model.predict(X_valid)
        r2 = r2_score(y_valid, y_pred)
        print(f"{model_name} - R2 Score on Test Data: {r2:.4f}")
        continue

    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_to_train, y_to_train)
    
    print(f"{model_name} - Best Parameters: {grid_search.best_params_}")
    print(f"{model_name} - Best Score: {-grid_search.best_score_:.4f}")
    
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_valid)
    r2 = r2_score(y_valid, y_pred)
    print(f"{model_name} - R2 score on Test Data: {r2:.4f}")
```
